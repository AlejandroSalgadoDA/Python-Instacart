```python
# Find the mean of orders by department id
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

This is the mean for all the department_ids in the dataset as opposed to the first one million rows. The dataset that shows one milliion rows is a smaller dataset and is not as accurate as the complete dataset. 

```python
# create a loyalty flag
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['customer_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
# create loyalty flag with parameters
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge.head()
```

```python
# Count number of customers by loyalty flag
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['customer_id', 'loyalty_flag', 'order_number']].head(20)
```

```python
# check the mean min max  
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean','min','max']})
```

```python
# transform prices to see the average price
ords_prods_merge['average_price']= ords_prods_merge.groupby(['customer_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(60)
```

```python
# create new columns 
ords_prods_merge.loc[ords_prods_merge['average_price']<10, 'spending_flag']= 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price']>=10, 'spending_flag']= 'High spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

```python
ords_prods_merge[['customer_id', 'spending_flag', 'loyalty_flag', 'order_number']].head(20)
```

```python
# create order frequency flag 
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['customer_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.shape
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge = ords_prods_merge.drop(columns = ['median_days_since_previous_order'])
```

```python
ords_prods_merge.head()
```

```python
# create frequency flag 
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order']>20, 'Frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order']>10) & (ords_prods_merge['median_days_since_prior_order']<=20), 'Frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order']<=10, 'Frequency_flag'] = 'Frequent customer'
```
